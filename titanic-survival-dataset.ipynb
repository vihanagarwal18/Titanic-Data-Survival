```python
#dropped these column from data frame
train_dataset = train_dataset.drop(['Ticket', 'Cabin'], axis=1)
train_dataset.head()
```

```python
train_dataset['Title']=train_dataset.Name.str.extract('([A-Za-z]+)\.',expand=False)
train_dataset['Title'].unique()
```

```python
pd.crosstab(train_dataset['Title'],train_dataset['Sex'])
```

```python
#col ,Dona ,Dr,Ms,rev are rare title
# so we will replace them with another keyword lets say 'rare'
train_dataset['Title']=train_dataset['Title'].replace(['Col','Dona','Dr','Ms','Rev'],'rare')
train_dataset['Title'].unique()
```

```python
train_dataset[['Title','Survived']].groupby('Title',as_index=False).mean()
```

```python
#now instead of writing title name we can write title id in our data frame 
title_mapping={"Mr":1,'Mrs':2,'Miss':3,'Master':4,'rare':5}
train_dataset['Title']=train_dataset['Title'].map(title_mapping)
train_dataset['Title']=train_dataset['Title'].fillna(0)
train_dataset
```

```python
train_dataset['Sex'] = train_dataset['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
type(train_dataset['Sex'][0])
#train_dataset
```

```python
guess_ages = np.zeros((2,3))
guess_ages
```

```python
for i in range (0,2):  # 2 because of 2 gender
    for j in range (0,3):  #3 beacause of 3 Pclass
        age_df=train_dataset[(train_dataset['Sex']==i) & (train_dataset['Pclass']==j+1)]['Age'].dropna()
        age_guess=age_df.median()
        guess_ages[i,j]=int(age_guess/0.5 +0.5)* 0.5   #basically x+ 0.25 banaya
guess_ages
```

```python
#now we will fill those ages with guess_ages wherver there is null
for i in range(0,2):
    for j in range (0,3):
        train_dataset.loc[(train_dataset.Age.isnull())&(train_dataset['Sex']==i) & (train_dataset['Pclass']==j+1),'Age']=guess_ages[i,j]
train_dataset['Age']=train_dataset['Age'].astype(int) 
train_dataset
```

```python
train_dataset['AgeBand'] = pd.cut(train_dataset['Age'], 5)
train_dataset[['AgeBand','Survived']].groupby(['AgeBand'],as_index=False).mean().sort_values(by='AgeBand',ascending=True)
```

```python
#now we are categorizing ages
train_dataset.loc[ train_dataset['Age'] <= 15.2, 'Age'] = 0
train_dataset.loc[(train_dataset['Age'] > 15.2) & (train_dataset['Age'] <= 30.4), 'Age'] = 1
train_dataset.loc[(train_dataset['Age'] > 30.4) & (train_dataset['Age'] <= 45.6), 'Age'] = 2
train_dataset.loc[(train_dataset['Age'] > 45.6) & (train_dataset['Age'] <= 60.8), 'Age'] = 3
train_dataset.loc[(train_dataset['Age']) > 60.8 , 'Age']=4
train_dataset
```

```python
train_dataset = train_dataset.drop(['AgeBand'], axis=1)
train_dataset
```

```python
train_dataset['FamilySize'] = train_dataset['SibSp'] + train_dataset['Parch'] + 1
train_dataset[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index=False).mean()
```

```python
train_dataset['IsAlone'] = 0
train_dataset.loc[train_dataset['FamilySize'] == 1, 'IsAlone'] = 1
train_dataset[['IsAlone', 'Survived']].groupby(['IsAlone'], as_index=False).mean()
```

```python
train_dataset = train_dataset.drop(['Parch', 'SibSp', 'FamilySize'], axis=1)
train_dataset.head()
```

```python
#calculating most frequent value in embarked data
freq_port = train_dataset.Embarked.dropna().mode()[0] # 0th index will give value with highest mode
freq_port
```

```python
train_dataset['Embarked'] = train_dataset['Embarked'].fillna(freq_port)
train_dataset[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean()
```

```python
train_dataset['Embarked'] = train_dataset['Embarked'].map({'S': 0,'C':1,'Q': 2}).astype(int)
train_dataset.head()
```

```python
#The inplace=True parameter ensures that this modification is done directly to the original DataFrame 
#rather than returning a modified copy.
train_dataset['Fare'].fillna(train_dataset['Fare'].dropna().median(), inplace=True)
train_dataset
```

```python
train_dataset['FareBand'] = pd.qcut(train_dataset['Fare'], 4)
train_dataset[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean()
```

```python
train_dataset.loc[ train_dataset['Fare'] <= 7.896, 'Fare'] = 0
train_dataset.loc[(train_dataset['Fare'] > 7.896) & (train_dataset['Fare'] <= 14.454), 'Fare'] = 1
train_dataset.loc[(train_dataset['Fare'] > 14.454) & (train_dataset['Fare'] <= 31.472), 'Fare']   = 2
train_dataset.loc[ train_dataset['Fare'] > 31.472, 'Fare'] = 3
train_dataset['Fare'] = train_dataset['Fare'].astype(int)
train_dataset = train_dataset.drop(['FareBand'],axis=1)    
train_dataset.head(20)
```

```python
#now will train our model

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score
```

```python
predictors = train_dataset.drop(['Survived', 'PassengerId','Name'], axis=1)
target = train_dataset["Survived"]
x_train, x_val, y_train, y_val = train_test_split(predictors, target, test_size = 0.22, random_state = 0)
#only 78% of data would be used for training model and 22% will be used for testing the data

'''
x_train: The feature variables for the training set.
x_val: The feature variables for the validation set.
y_train: The target variable for the training set.
y_val: The target variable for the validation set.
'''
regressor = LinearRegression()
regressor.fit(x_train, y_train)

# Predicting the Test set results
y_pred = regressor.predict(x_val)
y_pred
```

```python
input_prediction=regressor.predict([[2,0,3,2,2,1,1]])
# as our input_prediction had sex as 0 which means male he would die
accuracy=accuracy_score([0],input_prediction.round())
print("Accuracy:", int(accuracy*100),"%")
```

```python
accuracy = accuracy_score(y_val, y_pred.round())
print("Accuracy:", int(accuracy*100),"%")
```

```python
#now predicting with logistic regression/classification
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()
lr_model.fit(x_train, y_train)
```

```python
y_pred = lr_model.predict(x_train)
print("Prediction on training set:", y_pred)
```

```python
print("Accuracy on training set:", lr_model.score(x_val, y_val))
```
